````python
import ast

def convert_str_to_list_of_dicts(input_str):
    try:

        result_list = ast.literal_eval(input_str)

        if isinstance(result_list, list):

            if all(isinstance(item, dict) for item in result_list):
                return result_list
    except (SyntaxError, ValueError):
        pass

    return []


````

````python
from datetime import datetime

# Fonction pour convertir la date
def convert_date(date_str):
    # Diviser la chaîne en mots
    parts = date_str.split()

    # Extraire le jour, le mois, l'année et l'heure
    day = int(parts[1])
    month = {
        'يناير': 1, 'فبراير': 2, 'مارس': 3, 'أبريل': 4, 'ماي': 5, 'يونيو': 6,
        'يوليوز': 7, 'غشت': 8, 'شتنبر': 9, 'أكتوبر': 10, 'نونبر': 11, 'دجنبر': 12
    }[parts[2]]
    year = int(parts[3])
    time = parts[-1]

    # Construire la date et l'heure au format datetime
    date_time = datetime(year, month, day)

    # Ajouter l'heure
    time_parts = time.split(':')
    date_time = date_time.replace(hour=int(time_parts[0]), minute=int(time_parts[1]))

    return date_time
````

# Data Preprocessing
When preprocessing social media data from different sources, the output format can be standardized to include common fields. The standard format includes the following fields:

1. **Comment**: The actual text content of the comment.
2. **Date of Comment**: The timestamp indicating when the comment was made.
3. **Language**: The language of the comment (optional).
4. **source**: From where we get this data .
5. **Input**: Additional input or metadata related to the comment (optional).

````python
import re

def classify_language(comment):
    """
    Classify the language of a comment as 'ar' (Darija-arabic) or 'fr' (darija-French).
    """
    # Regular expressions for Arabic and French characters
    arabic_pattern = re.compile(r'[\u0600-\u06FF\u0750-\u077F\u08A0-\u08FF]+')  # Arabic Unicode range
    english_pattern = re.compile(r'[\x00-\x7F\x80-\xFF]+')  # English Unicode range

    # Check if the comment contains Arabic or French characters
    if arabic_pattern.search(comment):
        return 'ar'
    elif english_pattern.search(comment):
        return 'fr'
    else:
        return None  # Return None if no Arabic or French characters are found
````

````python
facebook_df['Language'] = facebook_df['comment_text'].apply(classify_language)
````

````python
import re

def clean_comment(comment):
    """
    Remove links and special characters from a comment using regular expressions.
    """
    # Regular expression pattern to match URLs
    url_pattern = re.compile(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')

    # Replace URLs and special characters with an empty string
    cleaned_comment = re.sub(url_pattern, '', comment)

    return cleaned_comment

# Apply cleaning to comments
df_youtube['Comment'] = df_youtube['Comment'].apply(clean_comment)
````

````python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification

def get_comment_sentiment(comment):
    # Load the DarijaBERT tokenizer and model
    tokenizer = AutoTokenizer.from_pretrained("SI2M-Lab/DarijaBERT")
    sentiment_model = AutoModelForSequenceClassification.from_pretrained("SI2M-Lab/DarijaBERT")

    # Tokenize the comment for sentiment analysis
    sentiment_inputs = tokenizer.encode_plus(
        comment,
        None,
        add_special_tokens=True,
        max_length=512,
        truncation=True,
        padding='longest',
        return_tensors='pt'
    )

    # Pass the input through the sentiment analysis model
    sentiment_outputs = sentiment_model(**sentiment_inputs)

    # Perform sentiment classification
    sentiment_logits = sentiment_outputs.logits
    sentiment = "Positive" if sentiment_logits[0][0] > sentiment_logits[0][1] else "Negative"

    return sentiment

# Usage example
comment = "مرحبا، كيف الحال؟"
sentiment = get_comment_sentiment(comment)
print('Sentiment:', sentiment)
````
