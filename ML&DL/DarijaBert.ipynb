# DarjaBERT Model :
 **is the first BERT model for the Moroccan Arabic dialect called “Darija”. It is based on the same architecture as BERT-base, but without the Next Sentence Prediction (NSP) objective. This model was trained on a total of ~3 Million sequences of Darija dialect representing 691MB of text or a total of ~100M tokens.**

**The model was trained on a dataset issued from three different sources:**

**Stories written in Darija scrapped from a dedicated website
Youtube comments from 40 different Moroccan channels**
**Tweets crawled based on a list of Darija keywords.**

```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification

def perform_sentiment_analysis(tweet_file, dialect_file):
    tweets_df = pd.read_csv(tweet_file)
    dialects_df = pd.read_json(dialect_file)

    # Load the DarijaBERT tokenizer and model
    tokenizer = AutoTokenizer.from_pretrained("SI2M-Lab/DarijaBERT")
    dialect_model = AutoModelForSequenceClassification.from_pretrained("SI2M-Lab/DarijaBERT")

    # Load the sentiment analysis tokenizer and model
    sentiment_tokenizer = AutoTokenizer.from_pretrained("SI2M-Lab/DarijaBERT")
    sentiment_model = AutoModelForSequenceClassification.from_pretrained("SI2M-Lab/DarijaBERT")

    dialects_df = dialects_df.reset_index(drop=True)

    results = []
    for index, row in tweets_df.iterrows():
        tweet = tweet = str(row['clean_comment'])

        if index >= len(dialects_df):
            print(f"Skipping tweet at index {index} due to missing dialect information.")
            continue

        dialect = dialects_df.iloc[index][0]

        dialect_inputs = tokenizer(tweet, return_tensors="pt")

        dialect_outputs = dialect_model(**dialect_inputs)

        sentiment_inputs = sentiment_tokenizer(tweet, return_tensors="pt")

        sentiment_outputs = sentiment_model(**sentiment_inputs)

        sentiment_logits = sentiment_outputs.logits
        sentiment = "Positive" if sentiment_logits[0][0] > sentiment_logits[0][1] else "Negative"

        result = {
            'Tweet': tweet,
            'Dialect': dialect,
            'Sentiment': sentiment,
        }
        results.append(result)

    results_df = pd.DataFrame(results)

    results_df.to_csv('/content/drive/MyDrive/sentiment_analysis_results.csv', index=False)
    print("Sentiment analysis completed. Results saved to sentiment_analysis_results.csv")

tweet_file = '/content/drive/MyDrive/DataSets sentiment analysis/finale_data/cleanforML_ar.csv'
dialect_file = '/content/drive/MyDrive/all_stop_words.json'

perform_sentiment_analysis(tweet_file, dialect_file)
```
